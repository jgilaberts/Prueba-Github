```python
import pandas as pd
```

# Leer Contenido

```python
datos = pd.read_csv('job_postings.csv')
datos
```

```python
docs = datos['description'] 
categs = datos['fraudulent'] 
```

# Obtener e imprimir matriz TFIDF

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
docs_bow = vectorizer.fit_transform(docs)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf = TfidfTransformer() 
docs_tfidf = tfidf.fit_transform(docs_bow)
docs_tfidf_densa = docs_tfidf.todense()
docs_tfidf_densa
```

# Entrenamiento Arbol de decision

```python
from sklearn.model_selection import train_test_split
docs_train, docs_test, categs_train, categs_test = train_test_split(docs_tfidf, categs, test_size = 0.25, 
                                                                    random_state = 50)
```

```python
from sklearn import tree

a_dec = tree.DecisionTreeClassifier()
a_dec.fit(docs_train, categs_train)

acc_train = a_dec.score(docs_train, categs_train)
acc_test = a_dec.score(docs_test, categs_test)

print("Accuracy entrenamiento: ", acc_train)
print("Accuracy PRUEBA: ", acc_test)
print("Fiabilidad: ", acc_test / acc_train)  
```

```python

from sklearn.metrics import confusion_matrix
categs_pred = a_dec.predict(docs_test)

cm = confusion_matrix(categs_test, categs_pred)
cm
```

### Podemos ver un accuracy de entrenamiento de 1, que comparandolo con el ACC de prueba, esta gran diferencia de 1 a 0.83, nos señala overfitting, con lo que deberiamos hacer el modelo menos complejo, por ejemplo usando un regularizador para que el modelo generalice mejor.

### Y luego viendo la matriz de confusion podemos ver que tenemos 71 Falsos Positivos y 53 Falsos Negativos, habria que trabajar en bajar estos falsos para obtener una mayor precision de nuestro modelo, seguramente con un regularizador como dicho anteriormente, estos valores bajarian.
